```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.random.randn(500) - 0.5
x1 = np.random.randn(500) - 0.5
print(f"{x0.shape}, {x1.shape}")

X = np.array(list(zip(x0, x1)))
print(X.shape)
```

```python
y = np.array([1 if i0 * i1 > 0 else 0 for (i0, i1)  in list(zip(x0, x1))])
print(y.shape)
```

```python
plt.scatter(
    X[:,0],
    X[:,1],
    c=y
)
```

```python
from sklearn import neighbors

knn = neighbors.KNeighborsClassifier(n_neighbors=1)
knn.fit(X=X, y=y)
knnFeat = knn.kneighbors(X=X, n_neighbors=1, return_distance=True)

knnFeat = np.array(list(knnFeat))
print(knnFeat.shape)
```
